```python
import pandas as pd
import matplotlib.pyplot as plt

# URLs das 4 lojas
urls = {
    "Loja 1": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv",
    "Loja 2": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv",
    "Loja 3": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv",
    "Loja 4": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"
}

# Calcular a média do frete de cada loja
medias_frete = {}
for loja, url in urls.items():
    df = pd.read_csv(url)
    medias_frete[loja] = df['Frete'].mean()

# Plotar gráfico de barras
plt.figure(figsize=(8,6))
bars = plt.bar(medias_frete.keys(), medias_frete.values(), color=['#4c72b0', '#55a868', '#c44e52', '#8172b3'])
plt.title('Comparação da Média do Custo do Frete das 4 Lojas')
plt.ylabel('Média do Frete (R$)')
plt.xlabel('Lojas')
plt.ylim(0, max(medias_frete.values()) * 1.2)

# Adicionar os valores das médias acima das barras
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.05, f'R${yval:.2f}', ha='center', va='bottom')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# URLs das 4 lojas
urls = {
    "Loja 1": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv",
    "Loja 2": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv",
    "Loja 3": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv",
    "Loja 4": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"
}

lucros = {}

for loja, url in urls.items():
    df = pd.read_csv(url)

    if 'Quantidade' in df.columns:
        total_vendas = (df['Preço'] * df['Quantidade']).sum()
    else:
        total_vendas = df['Preço'].sum()

    custo_frete = df['Frete'].sum()

    lucro = total_vendas - custo_frete
    lucros[loja] = lucro

# Gráfico de pizza
plt.figure(figsize=(8,8))
plt.pie(lucros.values(), labels=lucros.keys(), autopct='%1.1f%%', startangle=140, colors=plt.cm.Pastel1.colors)
plt.title('Participação Percentual no Lucro Total das Lojas')
plt.axis('equal')  # Para deixar o círculo perfeito
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# URLs das 4 lojas
urls = {
    "Loja 1": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv",
    "Loja 2": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv",
    "Loja 3": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv",
    "Loja 4": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"
}

# Dicionário para armazenar contagem de categorias por loja
categorias_lojas = {}

for loja, url in urls.items():
    df = pd.read_csv(url)
    contagem = df['Categoria do Produto'].value_counts()
    categorias_lojas[loja] = contagem

# Transformar em DataFrame
df_categorias = pd.DataFrame(categorias_lojas).fillna(0)

# Ordenar categorias pelo total geral
df_categorias['Total'] = df_categorias.sum(axis=1)
df_categorias = df_categorias.sort_values(by='Total', ascending=False).drop('Total', axis=1)

# Plotar gráfico de barras agrupadas
ax = df_categorias.plot(kind='bar', figsize=(12,7), colormap='tab20')
plt.title('Comparação da Distribuição das Categorias de Produtos nas 4 Lojas')
plt.ylabel('Quantidade Vendida')
plt.xlabel('Categoria do Produto')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Lojas')
plt.tight_layout()
plt.show()
```
